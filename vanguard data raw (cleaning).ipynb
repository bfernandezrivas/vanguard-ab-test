# Cleaning df_final_demo


```python
import pandas as pd
```

```python
df_final_demo = pd.read_csv('data raw/df_final_demo.txt')
```

```python
df_final_demo.head()
```

```python
df_final_demo.info()
```

```python
df_final_demo.describe().round(2)
```

### Renaming columns with clear names

```python
df_final_demo = df_final_demo.rename(columns={
    'client_id' : 'client_id',
    'clnt_tenure_yr': 'membership_years',
    'clnt_tenure_mnth': 'membership_months',
    'clnt_age' : 'client_age',
    'gendr'  : 'gender',
    'num_accts' : 'num_accounts',
    'bal' : 'total_account_balance',
    'calls_6_mnth' : 'num_calls_last_6_months',
    'logons_6_mnth' : 'num_logins_last_6_months'
    })
```

```python
df_final_demo.head()
```

### Cheking for nulls

```python
df_final_demo.isna().sum()
```

```python
cols_with_nulls = [
    'membership_years',
    'membership_months',
    'client_age',
    'gender',
    'num_accounts',
    'total_account_balance',
    'num_calls_last_6_months',
    'num_logins_last_6_months'
]

rows_with_nulls = df_final_demo[df_final_demo[cols_with_nulls].isna().any(axis=1)]
rows_with_nulls
```

### Removing rows with nulls

```python
cols_with_nulls = [
    'membership_years',
    'membership_months',
    'client_age',
    'gender',
    'num_accounts',
    'total_account_balance',
    'num_calls_last_6_months',
    'num_logins_last_6_months'
]

df_final_demo = df_final_demo.dropna(subset=cols_with_nulls)
```

```python
df_final_demo.isna().sum()

```

```python
df_final_demo.info()
```

```python
df_final_demo.sample(20)
```

### Saving clean csv file

```python
df_final_demo.to_csv('data clean/df_final_demo_clean.csv',index=False)
```

# Cleaning df_final_experiment_clients

```python
df_final_experiment_clients = pd.read_csv('data raw/df_final_experiment_clients.txt')
df_final_experiment_clients.head()
```

```python
df_final_experiment_clients.info()
```

### Dropping the 15 clients with null data from df_final_demo

```python
client_ids_to_remove = [
    7402828, 355337, 8412164, 4666211, 2222915,
    4876926, 5277910, 7616759, 8191345, 1227228,
    8611797, 5144725, 1037867, 1618749, 6021001
]

df_final_experiment_clients = df_final_experiment_clients[
    ~df_final_experiment_clients['client_id'].isin(client_ids_to_remove)
]
```

```python
df_final_experiment_clients.info()
```

```python
df_final_experiment_clients.nunique()
```

```python
df_final_experiment_clients['Variation'].unique()
```

### Renaming column "Variation"

```python
df_final_experiment_clients = df_final_experiment_clients.rename(columns={
    'client_id' : 'client_id',
    'Variation': 'experiment_group'
    })
```

```python
df_final_experiment_clients.head(10)
```

### Saving clean csv file

```python
df_final_experiment_clients.to_csv('data clean/df_final_experiment_clients_clean.csv',index=False)
```
